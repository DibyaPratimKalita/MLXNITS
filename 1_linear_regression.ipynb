```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mango.csv')
df
```

```python
%matplotlib inline
plt.xlabel('HistoryP')
plt.ylabel('price')
plt.scatter(df.HistoryP,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
# Create linear regression object
model = linear_model.LinearRegression()
model.fit(new_df,price)
```

**(1) Predict price of a home with area = 3300 sqr ft**

```python
model.predict([[33]])
```

```python
import joblib
joblib.dump(model, 'Mango.pkl')
```

```python
model = joblib.load('Mango.pkl')
test_prediction = model.predict([[1200]])  # Predict for an example area
print(test_prediction)
```
